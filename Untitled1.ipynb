```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os

# Create a folder named 'graphs' if it doesn't already exist.
# This ensures the script doesn't crash when trying to save the plots.
if not os.path.exists('graphs'):
    os.makedirs('graphs')

def bisection_method(func, xl, xu, max_iter=20):
    """
    Implements the Bisection Method to find the root of a function.
    
    Logic: It repeatedly divides the interval in half. If the function 
    changes sign over an interval, a root must exist (Intermediate Value Theorem).
    
    Args:
        func: The mathematical equation (lambda function).
        xl: The lower boundary of the search interval.
        xu: The upper boundary of the search interval.
        max_iter: Fixed number of iterations to perform.
    """
    # Check the 'Bracketing Condition': f(xl) and f(xu) must have opposite signs.
    # If the product is positive, the signs are the same, and the method won't work.
    if func(xl) * func(xu) >= 0:
        return None

    data = []      # List to store data for each iteration
    xm_old = None  # Variable to store the previous midpoint for error calculation

    for i in range(1, max_iter + 1):
        # 1. Calculate the midpoint (the current estimate of the root)
        xm = (xl + xu) / 2.0
        
        # 2. Calculate f(xm) to see where the curve is at the midpoint
        fxm = func(xm)
        
        # 3. Calculate Absolute Relative Approximate Error (%)
        # This tells us how much the root estimate is changing compared to the last step.
        if xm_old is not None and xm != 0:
            ea = abs((xm - xm_old) / xm) * 100
        else:
            ea = 0.0 # Error is zero for the first iteration
            
        # Append results to the list for later display in a table
        data.append({"Iter": i, "xl": xl, "xu": xu, "xm": xm, "Error (%)": ea})
        
        # 4. Determine which sub-interval contains the root
        # If f(xl)*f(xm) < 0, the root is in the left half. Update the upper bound (xu).
        if func(xl) * fxm < 0:
            xu = xm
        # Otherwise, the root is in the right half. Update the lower bound (xl).
        else:
            xl = xm
            
        # Store current xm as 'old' for the next iteration's error calculation
        xm_old = xm
        
    # Return the data as a Pandas DataFrame for easy table viewing
    return pd.DataFrame(data)

# Problem Definitions: A list of dictionaries containing equations and guesses.
# We use Python syntax: ** for exponents and * for multiplication.
problems = [
    {"id": "F1", "eqn": "x^5 + x - 1", "func": lambda x: x**5 + x - 1, "guess": [0, 1]},
    {"id": "F2", "eqn": "x^5 + 5x^3 - 4x + 1", "func": lambda x: x**5 + 5*x**3 - 4*x + 1, "guess": [-1, 0]},
    {"id": "F3", "eqn": "x^5 + 2x^4 - x - 3", "func": lambda x: x**5 + 2*x**4 - x - 3, "guess": [1, 2]},
    {"id": "F4", "eqn": "x^5 - 10", "func": lambda x: x**5 - 10, "guess": [1, 2]},
    {"id": "F5", "eqn": "x^5 + 2x^2 + x - 0.5", "func": lambda x: x**5 + 2*x**2 + x - 0.5, "guess": [0, 1]}
]

# --- Main Execution Loop ---
# This loop processes each equation one by one.
for p in problems:
    print(f"\nProcessing {p['eqn']}...")
    
    # Call the bisection function
    df = bisection_method(p["func"], p["guess"][0], p["guess"][1])
    
    if df is not None:
        # Display the numerical table in the console/output
        print(df[['Iter', 'xm', 'Error (%)']].to_string(index=False))
        
        # --- Visualization Logic ---
        # We grab the final root from the last row of our table
        root = df.iloc[-1]['xm']
        
        # Set a very small zoom range (0.05 units) around the root
        zoom = 0.05
        x_vals = np.linspace(root - zoom, root + zoom, 500)
        y_vals = p["func"](x_vals)
        
        # Generate the Plot
        plt.figure(figsize=(8, 5))
        plt.plot(x_vals, y_vals, label='f(x)', color='blue', linewidth=2)
        plt.axhline(0, color='black', lw=1)             # Draw the X-axis
        plt.axvline(root, color='red', linestyle='--', alpha=0.5) # Reference line at root
        plt.scatter([root], [0], color='red', zorder=5, label=f'Root  {root:.5f}')
        
        plt.title(f"Zoomed Root View: {p['eqn']}")
        plt.xlabel("x")
        plt.ylabel("f(x)")
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        # --- File Saving Logic for GitHub ---
        # This creates a unique filename for each problem inside the 'graphs' folder.
        filename = f"graphs/root_zoom_{p['id']}.png"
        plt.savefig(filename, dpi=300) # Save at high resolution (300 DPI)
        print(f"Graph saved locally as {filename}")
        
        plt.show() # Display the graph on screen
```

```python
import numpy as np               # Library for numerical calculations (arrays, linspace)
import matplotlib.pyplot as plt  # Library for creating charts and graphs
import pandas as pd              # Library for organizing data into nice tables
import os                        # Library for operating system commands (creating folders)

# --- CONFIGURATION ---
# Create a folder named 'graphs' to store the images. 
# This is best practice for GitHub so your repo stays organized.
if not os.path.exists('graphs'):
    os.makedirs('graphs')

# --- CORE ALGORITHM ---
def bisection_method(func, xl, xu, max_iter=20):
    """
    Implements the Bisection Method to find the root of a function.
    
    Parameters:
    - func: The mathematical function (e.g., lambda x: x**2 - 4)
    - xl:   The lower guess of the interval
    - xu:   The upper guess of the interval
    - max_iter: The number of times to run the loop (default is 20)
    """
    
    # 1. BRACKETING CHECK
    # We must check if the function changes sign between xl and xu.
    # If f(xl) * f(xu) is positive, they have the same sign (no root guaranteed).
    if func(xl) * func(xu) >= 0:
        print(f"Error: No root found between {xl} and {xu}.")
        return None

    results = []       # A list to store the history of our iterations
    xm_old = None      # To store the previous root estimate for error calculation

    # 2. ITERATION LOOP
    for i in range(1, max_iter + 1):
        # Calculate the Midpoint (xm) - This is our current best guess for the root
        xm = (xl + xu) / 2.0
        
        # Evaluate the function at the midpoint
        fxm = func(xm)
        
        # 3. CALCULATE ERROR (%)
        # Error = |(Current - Previous) / Current| * 100
        if xm_old is not None and xm != 0:
            ea = abs((xm - xm_old) / xm) * 100
        else:
            ea = 0.0  # Error is zero for the very first step
            
        # Add the data from this step to our results list
        results.append({
            "Iter": i,
            "xm (Root)": xm,
            "Error (%)": ea,
            "f(xm)": fxm
        })
        
        # 4. DECISION RULE (BISECTION LOGIC)
        # We need to decide which half of the interval to keep.
        # If f(xl) and f(xm) have opposite signs, the root is in the LEFT half.
        if func(xl) * fxm < 0:
            xu = xm
        # Otherwise, the root is in the RIGHT half.
        else:
            xl = xm
            
        # Update xm_old for the next loop's error calculation
        xm_old = xm

    # Return the results as a clean Pandas DataFrame (Table)
    return pd.DataFrame(results)

# --- PROBLEM DEFINITIONS ---
# We store all 5 equations in a list of dictionaries.
# This makes it easy to loop through them without rewriting code.
problems = [
    # Problem 1: Standard quintic
    {"id": "P1", "label": "x^5 + x - 1", "func": lambda x: x**5 + x - 1, "guess": [0, 1]},
    
    # Problem 2: Steep curve, negative root
    {"id": "P2", "label": "x^5 + 5x^3 - 4x + 1", "func": lambda x: x**5 + 5*x**3 - 4*x + 1, "guess": [-1, 0]},
    
    # Problem 3: Positive root near 1.1
    {"id": "P3", "label": "x^5 + 2x^4 - x - 3", "func": lambda x: x**5 + 2*x**4 - x - 3, "guess": [1, 2]},
    
    # Problem 4: Finding the 5th root of 10
    {"id": "P4", "label": "x^5 - 10", "func": lambda x: x**5 - 10, "guess": [1, 2]},
    
    # Problem 5: Small root dominated by lower order terms
    {"id": "P5", "label": "x^5 + 2x^2 + x - 0.5", "func": lambda x: x**5 + 2*x**2 + x - 0.5, "guess": [0, 1]}
]

# --- MAIN EXECUTION ---
for p in problems:
    print(f"\n--- Solving {p['label']} ---")
    
    # Run the bisection method
    df = bisection_method(p["func"], p["guess"][0], p["guess"][1])
    
    if df is not None:
        # Get the final estimated root from the last row of the table
        final_root = df.iloc[-1]['xm (Root)']
        print(f"Final Root found at: {final_root:.6f}")
        
        # --- PLOTTING LOGIC (ZOOMED) ---
        plt.figure(figsize=(8, 6))
        
        # 1. Define the Zoom Window
        # We take the root and look just 0.1 units to the left and right.
        # This "zooms in" to show the crossing clearly.
        zoom_range = 0.1
        x_vals = np.linspace(final_root - zoom_range, final_root + zoom_range, 200)
        y_vals = p["func"](x_vals)
        
        # 2. Plot the Function Curve
        plt.plot(x_vals, y_vals, color='blue', linewidth=2, label='f(x)')
        
        # 3. Draw Axis Lines
        plt.axhline(0, color='black', linewidth=1) # The X-axis (y=0)
        plt.axvline(final_root, color='red', linestyle='--', linewidth=1, label=f'Root  {final_root:.4f}')
        
        # 4. Highlight the Intersection Point
        plt.scatter([final_root], [0], color='red', s=100, zorder=5)
        
        # 5. Add Labels and Titles
        plt.title(f"Zoomed Graph: {p['label']}\nRoot  {final_root:.5f}")
        plt.xlabel("x")
        plt.ylabel("f(x)")
        plt.legend()
        plt.grid(True, linestyle=':', alpha=0.6)
        
        # 6. Save the graph to the folder
        filename = f"graphs/{p['id']}_zoomed.png"
        plt.savefig(filename, dpi=100)
        print(f"Graph saved to: {filename}")
        
        # Show the plot on screen
        plt.show()
```

```python
import numpy as np               # For numerical math
import matplotlib.pyplot as plt  # For plotting graphs
import pandas as pd              # For creating the tables
import os                        # For file management

# Ensure the 'graphs' folder exists for GitHub saving
if not os.path.exists('graphs'):
    os.makedirs('graphs')

# --- 1. THE BISECTION ALGORITHM ---
def bisection_method(func, xl, xu, max_iter=20):
    """
    Performs the Bisection Method for exactly 'max_iter' loops.
    Returns a DataFrame containing all 20 approximate roots.
    """
    # Safety Check: Do the guesses bracket a root?
    if func(xl) * func(xu) >= 0:
        print("Error: Initial guesses do not bracket the root.")
        return None

    results = []      # List to hold the 20 roots
    xm_old = None     # Used to calculate the error relative to the previous step

    # Loop exactly 20 times
    for i in range(1, max_iter + 1):
        # Calculate the new approximate root (Midpoint)
        xm = (xl + xu) / 2.0
        
        # Calculate the function value at this point
        fxm = func(xm)
        
        # Calculate Error (%) - How much did the root change?
        if xm_old is not None and xm != 0:
            ea = abs((xm - xm_old) / xm) * 100
        else:
            ea = 100.0 # First step has 100% uncertainty (or undefined)
            
        # Add this step's data to our list
        results.append({
            "Iter": i,
            "xl": xl,
            "xu": xu,
            "xm (Root)": xm,      # <--- This is one of the "20 Roots"
            "Error (%)": ea,
            "f(xm)": fxm
        })
        
        # DECISION: Which side is the root on?
        if func(xl) * fxm < 0:
            xu = xm   # Root is in the left half
        else:
            xl = xm   # Root is in the right half
            
        xm_old = xm

    # Convert the list into a nice table
    return pd.DataFrame(results)

# --- 2. DEFINE THE 5 PROBLEMS ---
problems = [
    {"eqn": "x^5 + x - 1", "func": lambda x: x**5 + x - 1, "guess": [0, 1]},
    {"eqn": "x^5 + 5x^3 - 4x + 1", "func": lambda x: x**5 + 5*x**3 - 4*x + 1, "guess": [-1, 0]},
    {"eqn": "x^5 + 2x^4 - x - 3", "func": lambda x: x**5 + 2*x**4 - x - 3, "guess": [1, 2]},
    {"eqn": "x^5 - 10", "func": lambda x: x**5 - 10, "guess": [1, 2]},
    {"eqn": "x^5 + 2x^2 + x - 0.5", "func": lambda x: x**5 + 2*x**2 + x - 0.5, "guess": [0, 1]}
]

# --- 3. MAIN EXECUTION LOOP ---
for p in problems:
    print(f"\n" + "="*60)
    print(f" SOLUTION FOR: {p['eqn']}")
    print("="*60)
    
    # Run the method
    df = bisection_method(p["func"], p["guess"][0], p["guess"][1])
    
    if df is not None:
        # PRINT THE TABLE OF 20 ROOTS
        # We format it to show 6 decimal places for precision
        pd.options.display.float_format = '{:.6f}'.format
        print(df[['Iter', 'xm (Root)', 'Error (%)', 'f(xm)']].to_string(index=False))
        
        # PLOT THE GRAPH (Zoomed in on the final root)
        final_root = df.iloc[-1]['xm (Root)']
        
        plt.figure(figsize=(8, 5))
        
        # Create a tiny window around the root (Zoom +/- 0.1)
        zoom = 0.1
        x_vals = np.linspace(final_root - zoom, final_root + zoom, 100)
        y_vals = p["func"](x_vals)
        
        plt.plot(x_vals, y_vals, 'b-', linewidth=2, label='f(x)')
        plt.axhline(0, color='black', linewidth=1)             # X-axis
        plt.axvline(final_root, color='red', linestyle='--')   # Vertical marker
        plt.scatter([final_root], [0], color='red', s=100, zorder=5) # The Root Dot
        
        plt.title(f"Final Root: {final_root:.5f} (After 20 Iterations)")
        plt.xlabel("x")
        plt.ylabel("f(x)")
        plt.grid(True, alpha=0.3)
        plt.legend()
        
        # Save and Show
        filename = f"graphs/root_{p['eqn'].replace(' ', '_')}.png"
        plt.savefig(filename)
        plt.show()
```
